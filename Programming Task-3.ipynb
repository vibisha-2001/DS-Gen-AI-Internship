```python
# Problem 1: Maximum Product of Two elements of an array
```

```python
def maxProduct(nums):
    max1 = 0
    max2 = 0
    for n in nums:
        if n > max1:
            max2 = max1
            max1 = n
        elif n > max2:
            max2 = n
    return (max1 - 1) * (max2 - 1)
print("maximum poduct of two elements:",maxProduct([3,4,5,2]))
```

```python
# Problem 2: Count Number of Teams
```

```python
def numTeams(rating):
    n = len(rating)
    count = 0
    
    for i in range(n):
        for j in range(i + 1, n):
            for k in range(j + 1, n):
                if rating[i] < rating[j] < rating[k]:
                    count += 1
                elif rating[i] > rating[j] > rating[k]:
                    count += 1
                    
    return count
print("Number of Teams:",numTeams([2,5,3,4,1]))
```

```python
# Problem 3: Number of Students Doing Homework at a Given Time
```

```python
def numStudents(start_time,end_time,query_time):
    start=np.array(start_time)
    end=np.array(end_time)
    active_students=(start<=query_time)&(end>=query_time)
    return np.sum(active_students)
print("Number of students:",numStudents([1,2,3],[3,2,7],4))
```

```python
# Problem 4: Number of Steps to Reduce a Number to Zero
```

```python
def numOfSteps(num):
    steps=0
    while num>0:
        if num%2==0:
            num//=2
        else:
            num-=1
        steps+=1
    return steps
print("Number of Steps:",numOfSteps(14))
```

```python
# Problem 5: Counting Bits
```

```python
def countBits(n):
    ans = []
    for i in range(n + 1):
        ans.append(bin(i).count('1'))
    return ans
print("count of bits:",countBits(2))
```
