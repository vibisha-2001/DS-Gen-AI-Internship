```python
# problem 1: Running sum of 1D array
```

```python
def running_sum(nums:list[int]):
    running_sum=[]
    sum=0
    for i in nums:
        sum+=i
        running_sum.append(sum)
    print("output:",running_sum)
running_sum([1,2,3,4,5])    
```

```python
# problem 2: Shuffle the Array 
```

```python
def shuffle(n:int,nums:list[int]):
    shuffled_array=[]
    for i in range(n):
        shuffled_array.append(nums[i])
        shuffled_array.append(nums[i+n])
    return shuffled_array
shuffle(3,[1,2,3,4,5,6])
```

```python
# problem 3: Kids with the Greatest Number of Candies
```

```python
def kids_with_candies(kid_candies:list[int],extra_candies:int):
    max_candies=max(kid_candies)
    output=[]
    for candies in kid_candies:
        if candies+extra_candies>=max_candies:
            output.append(True)
        else:
            output.append(False)
    return output
kids_with_candies([2,3,5,1,3],3)
```
