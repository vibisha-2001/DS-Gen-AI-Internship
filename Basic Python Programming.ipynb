```python
# problem 2: IF-Else, Weird or Not Weird
```

```python
n=int(input("Enter a number"))
print("input:",n)
if n%2!=0:
    print("Weird")
else:
    if n in range(2,6):
        print("Not Weird")
    elif n in range(6,21):
        print("Weird")
    else:
        print("Not Weird")
```

```python
# problem 6: Write a Function
```

```python
year=int(input("Enter a Year"))
print("input:",year)
def is_leap(year):
    if year%400==0:
        return True
    else:
        if year%4==0 and year%100!=0:
            return True
        else:
            return False
print(is_leap(year))
```
